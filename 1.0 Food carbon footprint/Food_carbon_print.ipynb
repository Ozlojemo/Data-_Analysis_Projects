```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Url="https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-02-18/food_consumption.csv"
data= pd.read_csv(Url)
print(data)
```

```python
# data_ea = data[data.country.isin(['Kenya','Uganda','Rwanda','Tanzania','Burundi']) & data.food_category.isin(['Beef','Pork','Lamb & Goat','Cheese'])]
data_ea = data[data.country.isin(['Kenya','Uganda','Rwanda','Tanzania','Burundi'])]
```
